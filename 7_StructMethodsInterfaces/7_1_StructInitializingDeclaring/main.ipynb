# Declaring and Initializing

## Declaring

```go
type <struct_name> struct {
	// list of fields
}
```

**fields**: group of variables

<ins>Ejemplos</ins>

```go
type Circle struct {
	x float64
	y float64
	r float64
}
```

```go
type Student struct {
	name string
	rollNo int
	marks []int
	grades map[string]int
}
```

## Initializing

### Method 1

```go
var <variable_name> <struct_name>
```

```go
var s Student
```

<ins>Ejemplo</ins>

````go
package main
import "fmt"

type Student struct {
	name string
	rollNo int
	marks []int
	grades map[string]int
}

func main(){
	var s Student
	fmt.Printf("%+v", s) // {ð’€¸name: ð’€¸rollNo:0 ð’€¸marks:[] ð’€¸grades:map[]}
}
main()
````

### Method 2

```go
<variable_name> := new(<struct_name>)
```

```go
s := new(Student)
```

````go
package main
import "fmt"

type Student struct {
	name string
	rollNo int
	marks []int
	grades map[string]int
}

func main(){
	s := new(Student)
	fmt.Printf("%+v", s) // &{ð’€¸name: ð’€¸rollNo:0 ð’€¸marks:[] ð’€¸grades:map[]}
}
main()
````

### Method 3

Iniciando con valores

```go
<variable_name> := <struct_name> {
	<field_name>: <value>,
	<field_name>: <value>,
}
```

```go
st := Student {
	name: "Diego",
	rollNo: 13,
}
```

````go
package main
import "fmt"

type Student struct {
	name string
	rollNo int
}

func main(){
	s := Student {
		name: "Diego",
		rollNo: 13,
	}
	fmt.Printf("%+v", s) // {ð’€¸name:Diego ð’€¸rollNo:13}

	st := Student { "Sergio", 14, }
	fmt.Printf("%+v", st) // {ð’€¸name:Sergio ð’€¸rollNo:14}
}
main()
````
